## EXTRACT DATA FROM IBB OPEN DATA PORTAL
*Connect to traffic & weather datasets and query with sql to get hourly aggregated data
*Store the results in pandas dataframe for further analysis

```python
#Import necessary modules
import requests
import pandas as pd
import json
from data_scrape import get_table_names, run_sql
```

```python
%load_ext sql
```

```python
#Method to get hourly data, update the dtypes from each table
def weather_agg(table_id):
    columns = run_sql('SELECT * FROM "TABLE_PLACEHOLDER" LIMIT 10', table_id, True)['id'].tolist()
    col_numeric = ''
    daily_avgs = ''
    not_99 = ''
    for col in columns[5:]:
        col_numeric+= f'"{col}"::numeric(9,2),'
        daily_avgs+= f'{col[:3].replace("E","G")}("{col}")::numeric(5,2) AS "{col}",'
        not_99 += f'"{col}"::numeric(9,2) != -99 AND "{col}"::numeric(9,2) != -8888 AND'
    col_numeric = col_numeric[:-1]     
    daily_avgs = daily_avgs[:-1]
    not_99 = not_99[:-5]
    
    sql_query = f'''
    WITH temp_table AS (
    SELECT DATE_TRUNC('hour', "DATE_TIME"::timestamp) AS "DATE", {col_numeric}
    FROM "TABLE_PLACEHOLDER" 
    WHERE {not_99}
    )

    SELECT "DATE", {daily_avgs}
    FROM temp_table
    GROUP BY "DATE"
    ORDER BY "DATE"
    '''
    return pd.DataFrame(run_sql(sql_query, table_id))
```

```python
weather_url = 'https://data.ibb.gov.tr/dataset/meteorology-observation-station-data-set'
tables = get_table_names(weather_url)
```

```python
#Extract data from all monthly tables and store in weather dataframe
for i,table_id in enumerate(tables): 
    if i == 0: 
        weather_agg(table_id).to_csv('weather_hourly_agg.csv', sep=',', index=False)
    else:
        weather_agg(table_id).to_csv('weather_hourly_agg.csv', sep=',', mode='a', header=False, index=False)
```

```python
#Method to get hourly data, update the dtypes from each table
def traffic_agg(table_id):
    columns = run_sql('SELECT * FROM "TABLE_PLACEHOLDER" LIMIT 10', table_id, True)['id'].tolist()
    col_numeric = ''
    daily_avgs = ''
    for col in columns[6:]:
        col_numeric+= f'"{col}"::integer,'  
        if col.startswith('NUM'):
            daily_avgs += f'MIN("{col}") AS MIN_{col},'
            daily_avgs += f'MAX("{col}") AS MAX_{col},'
            daily_avgs += f'AVG("{col}")::numeric(5,2) AS AVG_{col},'
        else:
            if col[:3] == 'AVE':
                daily_avgs+= f'{col[:3].replace("E","G")}("{col}")::numeric(5,2) AS "{col}",'
            else:
                daily_avgs+= f'{col[:3].replace("E","G")}("{col}") AS "{col}",'
    daily_avgs = daily_avgs[:-1]
    col_numeric = col_numeric[:-1]

    sql_query = f'''
    WITH temp_table AS (
    SELECT DATE_TRUNC('hour', "DATE_TIME"::timestamp) AS "DATE", {col_numeric}
    FROM "TABLE_PLACEHOLDER" 
    )

    SELECT "DATE", {daily_avgs}
    FROM temp_table
    GROUP BY "DATE"
    ORDER BY "DATE"
    '''
    
    return pd.DataFrame(run_sql(sql_query, table_id))
```

```python
traffic_url = 'https://data.ibb.gov.tr/dataset/hourly-traffic-density-data-set'
tables = get_table_names(traffic_url)
```

```python
#Extract data from all monthly tables and store in traffic dataframe
for i,table_id in enumerate(tables): 
    if i == 0: 
        traffic_agg(table_id).to_csv('traffic_hourly_agg.csv', sep=',', index=False)
    else:
        traffic_agg(table_id).to_csv('traffic_hourly_agg.csv', sep=',', mode='a', header=False, index=False)
```
